```python
from dotenv import load_dotenv

__ = load_dotenv("secrets.txt") # If key is in a file.
```

```python
from pydantic import BaseModel
from openai import AzureOpenAI
import os

client = AzureOpenAI(
    api_version = "2024-10-01-preview",
    api_key = os.getenv("AZURE_OPENAI_KEY"),
    azure_deployment = 'tts',
    azure_endpoint = os.getenv("AZURE_OPENAI_ENDPOINT")
)
```

```python

x = dict(
    model="tts-1",
    voice="alloy",
    input="Python and machine learning go together like snakes and ladders.",
)

with client.audio.speech.with_streaming_response.create(**x) as response:
    response.stream_to_file("speech.mp3")
```
